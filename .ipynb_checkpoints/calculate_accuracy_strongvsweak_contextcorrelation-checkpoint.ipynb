```python
data=pd.read_table("/media/noor/DataNS/Onderzoek/Projects/Backgroundtypicality/2018_scenecontext/model/resnet10_white_objectnoise.txt",delimiter="\t", names=('ind', 'prob', 'cat'))
print(data['ind'][0])


start = 'w_background/'
end = '/ILS'
data['corAns'] = data['ind'].str.extract('%s(.*)%s' % (start, end), expand=False)

data.ind = pd.to_numeric(data['ind'], errors='coerce').fillna(0)
data.corAns = pd.to_numeric(data['corAns'], errors='coerce').fillna(0)

correct_count = 0 

for i in range(0,(len(data)),6):
    #print(i)
    for ii in range(1,6):
        #print(data['corAns'][i])
        #print(data['ind'][i+ii])
        if data['corAns'][i]==data['ind'][i+ii]:
            #data['cor'][i]=1
            correct_count = correct_count + 1  
print(correct_count)
all_count = len(data)/6
print(len(data)/6)
acc_resnet10=[]
acc_resnet10.append((float(correct_count)/float(all_count)))
print(acc_resnet10)


data=pd.read_table("/media/noor/DataNS/Onderzoek/Projects/Backgroundtypicality/2018_scenecontext/model/resnet10_typical_objectnoise.txt",delimiter="\t", names=('ind', 'prob', 'cat'))
print(data['ind'][0])


start = 't_background/'
end = '/ILS'
data['corAns'] = data['ind'].str.extract('%s(.*)%s' % (start, end), expand=False)

data.ind = pd.to_numeric(data['ind'], errors='coerce').fillna(0)
data.corAns = pd.to_numeric(data['corAns'], errors='coerce').fillna(0)

correct_count = 0 

for i in range(0,(len(data)),6):
    #print(i)
    for ii in range(1,6):
        #print(data['corAns'][i])
        #print(data['ind'][i+ii])
        if data['corAns'][i]==data['ind'][i+ii]:
            #data['cor'][i]=1
            correct_count = correct_count + 1  
            
print(correct_count)
all_count = len(data)/6
print(len(data)/6)
acc_resnet10.append((float(correct_count)/float(all_count)))
print(acc_resnet10)


data=pd.read_table("/media/noor/DataNS/Onderzoek/Projects/Backgroundtypicality/2018_scenecontext/model/resnet10_atypical_objectnoise.txt",delimiter="\t", names=('ind', 'prob', 'cat'))
print(data['ind'][0])


start = 'a_background/'
end = '/ILS'
data['corAns'] = data['ind'].str.extract('%s(.*)%s' % (start, end), expand=False)

data.ind = pd.to_numeric(data['ind'], errors='coerce').fillna(0)
data.corAns = pd.to_numeric(data['corAns'], errors='coerce').fillna(0)

correct_count = 0 

for i in range(0,(len(data)),6):
    #print(i)
    for ii in range(1,6):
        #print(data['corAns'][i])
        #print(data['ind'][i+ii])
        if data['corAns'][i]==data['ind'][i+ii]:
            #data['cor'][i]=1
            correct_count = correct_count + 1  
            
print(correct_count)
all_count = len(data)/6
print(len(data)/6)
acc_resnet10.append((float(correct_count)/float(all_count)))
print(acc_resnet10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
import seaborn as sns
cmap = sns.color_palette("hls", 8)
sns.set_palette(cmap)
sns.set_style("white")
```

```python
plt.plot(acc_resnet10)
plt.plot(acc_resnet18)
plt.plot(acc_resnet34)
plt.plot(acc_resnet50)
plt.plot(acc_resnet101)
plt.plot(acc_resnet152)


plt.legend(['resnet10' ,'resnet18', 'resnet34','resnet50','resnet101', 'resnet152'])
#plt.legend(['low' , 'high'])
x = np.arange(0,6,1)
x_ticks_labels = ['white','typical','atypical']
axes = plt.axes()


# Set number of ticks for x-axis
axes.set_xticks(x)
# Set ticks labels for x-axis
axes.set_xticklabels(x_ticks_labels , fontsize=8)

plt.show()
```
